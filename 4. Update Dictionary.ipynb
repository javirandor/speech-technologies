# Updating dictionary

This code allows us to find all words in our transcription that are not included in the current language dictionary

```python
import os
import re
import unidecode
```

```python
fname = "/Users/javirando/Desktop/Universidad/3º/3 Trimestre/Speech/Final Project/models/es-es/corele.transcription"
count = 0
sentences = []
with open(fname, 'r') as f:
    for line in f:
        sentences.append(re.search(r'\<s\> (.*) \<\/s\>', line).group(1))
```

```python
sentences[0]
```

```python
dictname = "/Users/javirando/Desktop/Universidad/3º/3 Trimestre/Speech/Final Project/models/es-es/es.dict"
words_dict = []
with open(dictname, 'r') as f:
    for line in f:
        words_dict.append(re.search(r'(\w+) [\s\w]+', line).group(1))
```

```python
words_dict[0]
```

```python
missing = set()
```

```python
def find_missing(sentence, words_dict, missing):
    words = sentence.split(' ')
    return missing.union(set(words) - set(words_dict))
```

```python
missing = set()
for s in sentences:
    missing = find_missing(s, words_dict, missing)
```

```python
missing = list(missing)
```

```python
len(missing)
```

```python
# Includes the missing words in the dictionary (phonemes must be later manually included)
with open('/Users/javirando/Desktop/Universidad/3º/3 Trimestre/Speech/Final Project/models/es-es/es.dict', 'a') as f:
    for m in missing:
        if ('<' not in m):
            f.write(m+'\n')
```

```python
# Writes non-recognized words in the noise dictionary
with open('/Users/javirando/Desktop/Universidad/3º/3 Trimestre/Speech/Final Project/models/es-es/es-es/noisedict', 'a') as f:
    for m in missing:
        if ('<' not in m):
            f.write(m+' SIL\n')
```
